```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression, DecisionTreeRegressor, RandomForestRegressor, GBTRegressor
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml.linalg import Vectors
from pyspark.ml.classification import RandomForestClassifier
import pandas as pd
from pyspark.sql import functions as F
from pyspark.sql.functions import col, avg, date_format, concat, lpad
import matplotlib.pyplot as plt
import seaborn as sns

spark = SparkSession.builder \
    .appName("ML data training") \
    .getOrCreate()
```

```python
df = spark.read.csv("unclean_data.csv", header=True, inferSchema=True)

# Get the list of numeric columns
numeric_columns = [field.name for field in df.schema.fields if field.dataType.typeName() in ['double', 'float', 'int', 'long']]

# Calculate skewness for all numeric columns
skewness_results = {column: df.select(F.skewness(column)).first()[0] for column in numeric_columns}

# Convert results to a DataFrame for better visualization (optional)
skewness_df = spark.createDataFrame(skewness_results.items(), ["Column", "Skewness"])

# Show skewness values
skewness_df.show()
```

```python
numeric_columns = [field.name for field in df.schema.fields if field.dataType.typeName() in ['double', 'float', 'int', 'long']]

# Calculate the correlation matrix
correlation_matrix = df.select(numeric_columns).toPandas().corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Create a heatmap using seaborn
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5)

# Add titles and labels
plt.title('Correlation Matrix', fontsize=15)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
target_column = 'T2M'  # Example: Temperature at 2 meters
columns_of_interest = ['YEAR', 'MO', 'DY', 'WS50M', 'TS', 'WD50M', 'ALLSKY_KT']

# Calculate correlation coefficients with respect to the target column
correlations = {col: df.stat.corr(target_column, col) for col in columns_of_interest}

# Convert the correlation dictionary to lists for plotting
features = list(correlations.keys())
values = list(correlations.values())

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.barh(features, values, color='b')
plt.xlabel('Correlation Coefficient', fontsize=12)
plt.title(f'Correlation with {target_column}', fontsize=15)
plt.axvline(0, color='gray', linewidth=0.8, linestyle='--')  # Line at 0 for reference
plt.xlim(-1, 1)  # Set x-axis limits
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
df1 = df.withColumn("YEAR", col("YEAR").cast("int")) \
       .withColumn("MO", col("MO").cast("int")) \
       .withColumn("DY", col("DY").cast("int"))

# Group by YEAR and MO (month) and calculate the average for WS50M and WD50M
monthly_average = df1.groupBy("YEAR", "MO").agg(
    F.avg("WS50M").alias("avg_WS50M"),
    F.avg("WD50M").alias("avg_WD50M")
)

# Define a function to remove outliers based on IQR
def remove_outliers(df, col_name):
    # Calculate Q1 and Q3
    quantiles = df.approxQuantile(col_name, [0.25, 0.75], 0.01)
    q1 = quantiles[0]
    q3 = quantiles[1]
    iqr = q3 - q1
    
    # Define bounds
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Show bounds for debugging
    print(f"Bounds for {col_name} - Lower: {lower_bound}, Upper: {upper_bound}")
    
    # Filter out the outliers
    return df.filter((col(col_name) >= lower_bound) & (col(col_name) <= upper_bound))

# Remove outliers for WS50M and show counts
print("Count before removing WS50M outliers:", df1.count())
filtered_ws50m = remove_outliers(monthly_average, "avg_WS50M")
print("Count after removing WS50M outliers:", filtered_ws50m.count())

# Remove outliers for WD50M and show counts
print("Count before removing WD50M outliers:", df1.count())
filtered_wd50m = remove_outliers(filtered_ws50m, "avg_WD50M")
print("Count after removing WD50M outliers:", filtered_wd50m.count())

# Calculate new averages after outlier removal
final_monthly_average = filtered_wd50m.groupBy("YEAR", "MO").agg(
    F.avg("avg_WS50M").alias("avg_WS50M"),
    F.avg("avg_WD50M").alias("avg_WD50M")
)

# Order by YEAR and MO for better readability
final_monthly_average = final_monthly_average.orderBy("YEAR", "MO")

# Show the results
final_monthly_average.show()
```

```python
df_before_cleaning = monthly_average.toPandas()
df_after_cleaning = final_monthly_average.toPandas()

# Create a figure with 2 rows and 2 columns (before and after cleaning for WS50M and WD50M)
fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharex=True)

# Dot plot for WS50M before cleaning
sns.stripplot(x='MO', y='avg_WS50M', data=df_before_cleaning, ax=axes[0, 0], color='orange')
axes[0, 0].set_title('WS50M Before Cleaning')
axes[0, 0].set_ylabel('avg_WS50M (Wind Speed 50m)')

# Dot plot for WS50M after cleaning
sns.stripplot(x='MO', y='avg_WS50M', data=df_after_cleaning, ax=axes[1, 0], color='green')
axes[1, 0].set_title('WS50M After Cleaning')
axes[1, 0].set_ylabel('avg_WS50M (Wind Speed 50m)')

# Dot plot for WD50M before cleaning
sns.stripplot(x='MO', y='avg_WD50M', data=df_before_cleaning, ax=axes[0, 1], color='orange')
axes[0, 1].set_title('WD50M Before Cleaning')
axes[0, 1].set_ylabel('avg_WD50M (Wind Direction 50m)')

# Dot plot for WD50M after cleaning
sns.stripplot(x='MO', y='avg_WD50M', data=df_after_cleaning, ax=axes[1, 1], color='green')
axes[1, 1].set_title('WD50M After Cleaning')
axes[1, 1].set_ylabel('avg_WD50M (Wind Direction 50m)')

# Set common x-axis label for months
for ax in axes[1, :]:
    ax.set_xlabel('Month')

# Display the plots
plt.tight_layout()
plt.show()
```

```python
if "avg_WS50M" not in final_monthly_average.columns:
    print("avg_WS50M column is missing in final_monthly_average DataFrame.")
else:
    # Assemble features into a single vector
    assembler = VectorAssembler(inputCols=["avg_WS50M", "avg_WD50M"], outputCol="features")
    assembled_df = assembler.transform(final_monthly_average)

    # Split data into training and testing sets
    train_data, test_data = assembled_df.randomSplit([0.8, 0.2], seed=42)

    # List of models to evaluate
    models = {
        "Linear Regression": LinearRegression(featuresCol="features", labelCol="avg_WS50M"),  # Change label column here
        "Decision Tree": DecisionTreeRegressor(featuresCol="features", labelCol="avg_WS50M"),
        "Random Forest": RandomForestRegressor(featuresCol="features", labelCol="avg_WS50M", numTrees=100),
        "Gradient Boosting": GBTRegressor(featuresCol="features", labelCol="avg_WS50M")
    }

    # Initialize evaluator
    evaluator = RegressionEvaluator(labelCol="avg_WS50M", predictionCol="prediction")

    # Dictionary to store model performance
    performance = {}

    # Loop through models, fit them, and evaluate
    for model_name, model in models.items():
        # Fit model
        trained_model = model.fit(train_data)
        
        # Make predictions
        predictions = trained_model.transform(test_data)
        
        # Evaluate using RMSE and R
        rmse = evaluator.evaluate(predictions, {evaluator.metricName: "rmse"})
        r2 = evaluator.evaluate(predictions, {evaluator.metricName: "r2"})
        
        # Store performance metrics
        performance[model_name] = {
            "RMSE": rmse,
            "R": r2,
            "Accuracy Percentage": r2 * 100
        }

    # Print performance results
    for model_name, metrics in performance.items():
        print(f"{model_name}: RMSE = {metrics['RMSE']:.4f}, R = {metrics['R']:.4f}, Accuracy Percentage = {metrics['Accuracy Percentage']:.2f}%")
```

```python
rf = RandomForestRegressor(labelCol="avg_WS50M", featuresCol="features", numTrees=100)
  
# Fit the model on the training data
rf_model = rf.fit(train_data)

# Evaluate on the training data
train_predictions = rf_model.transform(train_data)
train_evaluator = RegressionEvaluator(labelCol="avg_WS50M", predictionCol="prediction", metricName="r2")
train_r2 = train_evaluator.evaluate(train_predictions)

# Evaluate on the test data
test_predictions = rf_model.transform(test_data)
test_evaluator = RegressionEvaluator(labelCol="avg_WS50M", predictionCol="prediction", metricName="r2")
test_r2 = test_evaluator.evaluate(test_predictions)

# Compare training and test R scores
print(f"Training R: {train_r2:.4f}")
print(f"Test R: {test_r2:.4f}")

# Check for overfitting
if train_r2 > test_r2:
    print("Potential overfitting detected. Training performance is better than test performance.")
else:
    print("No overfitting detected.")
```
